```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
total_resp = data['Country'].value_counts()
total_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
python_data = pd.concat([total_resp, total_python], axis='columns')
python_data
```

```python
python_data.rename(columns={'Country': 'Total Responders', 'LanguageWorkedWith': 'Knows Python'}, inplace=True)
python_data['% knows Python'] = (python_data['Knows Python'] / python_data['Total Responders']) * 100
python_data
```

```python
python_data.sort_values(by='% knows Python', ascending=False, inplace=True)
python_data.head(50)
```
